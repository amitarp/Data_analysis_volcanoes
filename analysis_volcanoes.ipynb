```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=pd.read_csv('Volcanoes.csv')
```

### Region having the highest number of volcanoes

```python
x=data['Region'].value_counts().index.to_list()
```

```python
y=data['Region'].value_counts().values.tolist()
```

```python
plt.figure(figsize=(15,12))
bar = sns.barplot(x =y,y = x,color = 'yellow',edgecolor = 'red')
plt.title('Region of volcanoes')
plt.xlabel('Count')
plt.ylabel('Regions')
plt.show()
```

#### Conclusion: South america has the most volcanoes followed by Indonesia

## Countries having the highest number of volcanoes

```python
value = data['Country'].value_counts().head(20)
x = value.index
y = value.values
```

```python
plt.figure(figsize=(25,12))
bar = sns.barplot(x =y,y = x,color = 'yellow',edgecolor = 'red')
plt.title('Countries of volcanoes')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

## Distribution of Elevation

```python
plt.figure(figsize=(25,12))
sns.histplot(x=data['Elevation (m)'],color='yellow',edgecolor='red')
plt.title('Distribution of Elevation')
plt.show()
```

### Conclusion: Most of the volcanoes have height between 200 to 2200 m

## Locations having the highest number of volcanoes

```python
labels=data['Location'].value_counts().head(5).keys()
explode=(0.2,0.1,0.1,0.1,0.1)
colors = ['#F2003C','#ec5300','#ec9b00','#ecca00','#FF6E4A']
plt.figure(figsize=(15,12))
plt.pie(x=data['Location'].value_counts().head(5),labels=labels,explode=explode,shadow=True,colors=colors)
plt.title("Location of volcanoes")
plt.show()
```

## Analysis on Types of Volcanoes

```python
x=data['Type'].value_counts().index.tolist()
y=data['Type'].value_counts().values.tolist()
```

```python
plt.figure(figsize=(15,12))
bar = sns.barplot(x =y,y = x,color = 'yellow',edgecolor = 'red')
plt.title('Types of volcanoes')
plt.xlabel('Count')
plt.ylabel('Types')
plt.show()
```

# Analysis on status of volcanoes

```python
labels = data['Status'].value_counts().head(5).keys()
explode = (0.2,0.1,0.1,0.1,0.1)
def my_fmt(x):
    return '{:.4f}%\n({:.0f})'.format(x, total*x/100)
colors = ['#F2003C','#ec5300','#ec9b00','#ecca00','#FF6E4A']
total = len(data['Status'].value_counts())
plt.figure(figsize=(15,12))
plt.pie(x = data['Status'].value_counts().head(5),
        labels = labels,
        explode = explode,
        shadow = True,
       autopct=my_fmt,
       colors = colors)
plt.show()
```
